### (d) Create the following list of months:

The output generated by your loop should look like this:

`"August is NOT part of the Spring semester"`

`"February is part of the Spring semester"`

- **Hint**: Use if/else statements inside the loop. You can use the command "in" (see Lecture 4) to test whether an element belongs to a list


```python

# Define the lists of months and the spring semester months
list_months = ["August", "February"]
list_spring_semester = ["January", "February", "March", "April", "May"]


for list_months in list_months:
    if list_months in list_spring_semester:
        print(f"{list_months} is part of the Spring semester")
    else:
        print(f"{list_months} is NOT part of the Spring semester")



```

### (f) Run loop through different datasets and compute statistics

```python
# Write your answer here:

wdi_ageprop_1980 = pd.read_csv("/Users/carolkao/Documents/GitHub/qtm151/quizzes/data_quiz/wdi_ageprop_1980.csv")
wdi_ageprop_2000 = pd.read_csv("/Users/carolkao/Documents/GitHub/qtm151/quizzes/data_quiz/wdi_ageprop_2000.csv")
wdi_ageprop_2020 = pd.read_csv("/Users/carolkao/Documents/GitHub/qtm151/quizzes/data_quiz/wdi_ageprop_2020.csv")


# List of datasets already loaded into memory
list_datasets = [wdi_ageprop_1980, wdi_ageprop_2000, wdi_ageprop_2020]

# For loop to compute the mean of percent_ages0to14
for data in list_datasets:
    mean_value = data["percent_ages0to14"].mean()
    print(f"The mean of percent_ages0to14 is: {mean_value}")


```
